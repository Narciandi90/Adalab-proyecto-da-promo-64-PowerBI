# EDA y limpieza inicial del dataset OSMI

Este notebook recoge una **revisión exploratoria básica** y una **limpieza inicial** del dataset de salud mental en el entorno laboral, con el objetivo de dejar una versión consistente y lista para su explotación posterior en **Power BI**.

## Objetivos del notebook
- Cargar el dataset original.
- Revisar su estructura general, nulos, duplicados y tipos de variables.
- Aplicar una limpieza básica coherente con el análisis.
- Realizar una limpieza final orientada a exportación.
- Generar un CSV limpio para la fase de visualización.

> **Nota:** este notebook no busca profundizar en un EDA avanzado, sino documentar de forma clara y ordenada la preparación del dato utilizada en el proyecto.


## 1. Imports y configuración

En esta sección se importan las librerías necesarias para la exploración y transformación del dataset.  
También se aplican algunos ajustes visuales para facilitar la lectura de tablas y salidas.

```python
# ==============================
# 📦 IMPORTS
# ==============================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración opcional para que se vea mejor
pd.set_option("display.max_columns", None)
plt.style.use("default")
```

## 2. Carga del dataset

Se carga el archivo CSV original desde la carpeta de datos en bruto y se realiza una comprobación rápida de dimensiones y primeras filas.

```python
# ==============================
# 📂 CARGA DEL DATASET
# ==============================

ruta = "../data/raw/survey.csv"

df = pd.read_csv(ruta)

print("✅ Dataset cargado correctamente")
print("Filas:", df.shape[0])
print("Columnas:", df.shape[1])

display(df.head())
```

## 3. Función de EDA básico

Se define una función reutilizable para obtener una visión general del dataset:
- dimensiones,
- tipos de datos,
- valores nulos,
- duplicados,
- estadísticas descriptivas,
- y distribución básica de variables categóricas.

```python
# ==============================
# 🔎 FUNCIÓN EDA BÁSICA
# ==============================

def eda_basico(df):
    """
    Realiza una exploración básica de un DataFrame.

    Parameters
    ----------
    df : pandas.DataFrame
        DataFrame sobre el que se quiere realizar la revisión exploratoria.

    Returns
    -------
    None
        La función no devuelve ningún valor. Imprime y muestra resúmenes
        descriptivos del dataset en pantalla.
    """
    
    print("📊 INFORMACIÓN GENERAL")
    print("-" * 50)
    print("Shape:", df.shape)
    print("\nTipos de datos:")
    display(df.dtypes)
    
    
    print("\n" + "="*60)
    print("🔍 VALORES NULOS")
    print("-" * 50)
    nulos = df.isnull().sum()
    porcentaje_nulos = (nulos / len(df)) * 100
    
    tabla_nulos = pd.DataFrame({
        "nulos": nulos,
        "% nulos": porcentaje_nulos
    }).sort_values(by="% nulos", ascending=False)
    
    display(tabla_nulos[tabla_nulos["nulos"] > 0])
    
    
    print("\n" + "="*60)
    print("🔁 DUPLICADOS")
    print("-" * 50)
    duplicados = df.duplicated().sum()
    print("Número de duplicados:", duplicados)
    
    
    print("\n" + "="*60)
    print("📈 VARIABLES NUMÉRICAS")
    print("-" * 50)
    display(df.describe())
    
    
    print("\n" + "="*60)
    print("🗂️ VARIABLES CATEGÓRICAS")
    print("-" * 50)
    
    columnas_categoricas = df.select_dtypes(include="object").columns
    
    for col in columnas_categoricas:
        print(f"\n🔹 {col}")
        print("Valores únicos:", df[col].nunique())
        display(df[col].value_counts().head())
```

## 4. Ejecución del EDA sobre el dataset original

Se ejecuta la función anterior sobre el dataset sin transformar para identificar rápidamente posibles problemas de calidad del dato.

```python
eda_basico(df)
```

## 5. Función de limpieza inicial

Aquí se aplica una limpieza básica orientada a homogeneizar el dataset antes de su uso en Power BI:
- eliminación de columnas poco útiles,
- filtrado de edades incoherentes,
- normalización de género,
- agrupación de países minoritarios,
- y eliminación de variables no prioritarias para el análisis.

```python
# ==============================
# 🧹 FUNCIÓN LIMPIEZA DATASET
# ==============================

def limpiar_dataset(df):
    """
    Aplica una limpieza inicial al dataset OSMI.

    Transformaciones principales:
    - Elimina columnas irrelevantes para el análisis.
    - Filtra edades fuera del rango considerado válido.
    - Estandariza la variable de género.
    - Agrupa países poco frecuentes en la categoría 'Other'.
    - Elimina la columna state por su baja utilidad en este contexto.

    Parameters
    ----------
    df : pandas.DataFrame
        DataFrame original.

    Returns
    -------
    pandas.DataFrame
        DataFrame limpio tras aplicar las transformaciones definidas.
    """

    df = df.copy()

    # eliminar columnas que no aportan al análisis
    df = df.drop(columns=["Timestamp", "comments"], errors="ignore")

    # limpiar edades incorrectas
    df = df[(df["Age"] >= 18) & (df["Age"] <= 65)]

    # normalizar gender
    df["Gender"] = df["Gender"].str.lower()

    gender_map = {
        "male": "Male",
        "m": "Male",
        "male-ish": "Male",
        "maile": "Male",
        "mal": "Male",
        "msle": "Male",

        "female": "Female",
        "f": "Female",
        "woman": "Female",
        "female ": "Female"
    }

    df["Gender"] = df["Gender"].replace(gender_map)

    # todo lo demás como Other
    df["Gender"] = df["Gender"].apply(
        lambda x: x if x in ["Male", "Female"] else "Other"
    )

    # agrupar países menos frecuentes
    top_countries = ["United States", "United Kingdom", "Canada", "Germany", "Netherlands"]

    df["Country"] = df["Country"].apply(
        lambda x: x if x in top_countries else "Other"
    )

    # eliminar state (solo aplica a USA)
    df = df.drop(columns=["state"], errors="ignore")

    print("✅ Limpieza completada")
    print("Filas:", df.shape[0])
    print("Columnas:", df.shape[1])

    return df
```

## 6. Aplicación de la limpieza inicial

Se genera una copia limpia del dataset original tras aplicar las transformaciones definidas en la función anterior.

```python
df_clean = limpiar_dataset(df)
```

## 7. Revisión rápida del dataset limpio

Se vuelve a ejecutar el EDA básico para comprobar el efecto de la limpieza inicial y validar que la estructura del dataset es coherente.

```python
eda_basico(df_clean)
```

## 8. Limpieza final antes de exportación

En esta fase se realizan los últimos ajustes necesarios para dejar el dataset preparado para visualización:
- eliminación de duplicados,
- imputación de nulos en variables clave,
- y revisión de nulos restantes.

```python
# ==============================
# 🧹 LIMPIEZA FINAL ANTES DE EXPORTAR
# ==============================

def limpieza_final_osmi(df):
    """
    Realiza la limpieza final del dataset antes de su exportación.

    Transformaciones principales:
    - Elimina registros duplicados.
    - Imputa nulos en `self_employed` con la moda.
    - Imputa nulos en `work_interfere` con una categoría explícita.
    - Muestra un pequeño resumen final de calidad del dato.

    Parameters
    ----------
    df : pandas.DataFrame
        DataFrame previamente depurado.

    Returns
    -------
    pandas.DataFrame
        DataFrame final listo para exportación y uso en Power BI.
    """
    df = df.copy()

    # 1) Eliminar duplicados
    duplicados_antes = df.duplicated().sum()
    df = df.drop_duplicates()
    duplicados_despues = df.duplicated().sum()

    # 2) Imputar self_employed con la moda
    if "self_employed" in df.columns:
        moda_self = df["self_employed"].mode(dropna=True)[0]
        df["self_employed"] = df["self_employed"].fillna(moda_self)

    # 3) Imputar work_interfere con nueva categoría
    if "work_interfere" in df.columns:
        df["work_interfere"] = df["work_interfere"].fillna("No response")

    # 4) Resumen rápido
    print("✅ Limpieza final completada")
    print(f"Duplicados antes: {duplicados_antes} | después: {duplicados_despues}")
    print("\nNulos restantes (solo >0):")
    nulos_restantes = df.isnull().sum()
    display(nulos_restantes[nulos_restantes > 0])

    return df
```

## 9. Dataset final preparado

Se crea la versión final del dataset que será exportada y consumida posteriormente en Power BI.

```python
df_final = limpieza_final_osmi(df_clean)
```

## 10. Exportación del CSV limpio

Se exporta el dataset procesado a la carpeta correspondiente para su uso en el flujo analítico y de visualización.

```python
# ==============================
# 📤 EXPORTAR CSV LIMPIO PARA POWER BI
# ==============================

ruta_salida = "../data/processed/osmi_mental_health_clean.csv"

df_final.to_csv(ruta_salida, index=False)

print("✅ Exportado correctamente")
print("Ruta:", ruta_salida)
```
